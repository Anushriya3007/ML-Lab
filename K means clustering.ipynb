```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Load Iris dataset directly
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# 2. Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['cluster'] = kmeans.fit_predict(df)

# 3. 2D scatter plot (Petal Length vs Petal Width)
plt.figure(figsize=(8, 6))
plt.scatter(df.iloc[:, 2], df.iloc[:, 3], c=df['cluster'], cmap='viridis', s=100, edgecolor='black')
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-Means Clustering (2D Plot)")
plt.colorbar(label="Cluster")
plt.grid(True)
plt.show()

# 4. 3D scatter plot (Sepal Length, Sepal Width, Petal Length)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2], c=df['cluster'], cmap='viridis', s=80, edgecolor='black')
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
ax.set_title("K-Means Clustering (3D Plot)")
fig.colorbar(scatter, ax=ax, label="Cluster")
plt.show()
```
