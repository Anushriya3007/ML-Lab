```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris # Import load_iris

# ---------------------------------------------------
# 1. Load the Iris dataset
# ---------------------------------------------------
# Load the Iris dataset directly from sklearn
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target] # Add species column for consistency if it's expected to be dropped

print("Original Dataset:")
print(df.head())

# ---------------------------------------------------
# 2. Remove the species (target) column
# ---------------------------------------------------
if "species" in df.columns:
    df_features = df.drop(columns=["species"])
    print("\nRemoved target column: species")
else:
    raise ValueError("Column 'species' not found. Tell me the correct column name.")

print("\nDataset WITHOUT species column:")
print(df_features.head())

# ---------------------------------------------------
# 3. Apply K-Means Clustering (3 clusters)
# ---------------------------------------------------
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')

# Fit the model
clusters = kmeans.fit_predict(df_features)

# Add cluster labels to dataframe
df_features["cluster"] = clusters

print("\nDataset WITH cluster labels:")
print(df_features.head())

# ---------------------------------------------------
# 4. Print Cluster Centers
# ---------------------------------------------------
print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```
